```python
import json
 
# Opening JSON file
f = open('users.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
```

```python
level1_0T = 0
level1_00 = 0
level1_01 = 0

level1_1T = 0
level1_10 = 0
level1_11 = 0

level1_2T = 0
level1_20 = 0
level1_21 = 0


level1_3T = 0
level1_30 = 0
level1_31 = 0

level1_4T = 0
level1_40 = 0
level1_41 = 0

level2_00 = 0
level2_01 = 0

level2_10 = 0
level2_11 = 0


level3_00 = 0
level3_01 = 0

level3_10 = 0
level3_11 = 0


for i in range(20):
    
    levels = data[i]['level']
    print(levels['2'][0]['initialGrade'])
    
    level1_00 += levels['1'][0]['initialGrade']
    level1_01 += levels['1'][0]['finalGrade']
    level1_0T += levels['1'][0]['timeSpent']
    
    level1_10 += levels['1'][1]['initialGrade']
    level1_11 += levels['1'][1]['finalGrade']
    level1_1T += levels['1'][1]['timeSpent']
    
    level1_20 += levels['1'][2]['initialGrade']
    level1_21 += levels['1'][2]['finalGrade']
    level1_2T += levels['1'][2]['timeSpent']
    
    level1_30 += levels['1'][3]['initialGrade']
    level1_31 += levels['1'][3]['finalGrade']
    level1_3T += levels['1'][3]['timeSpent']
    
    level1_40 += levels['1'][4]['initialGrade']
    level1_41 += levels['1'][4]['finalGrade']
    level1_4T += levels['1'][4]['timeSpent']
    
    level2_00 += levels['2'][0]['initialGrade']
    level2_01 += levels['2'][0]['finalGrade']

    level2_10 += levels['2'][1]['initialGrade']
    level2_11 += levels['2'][1]['finalGrade']
    
    
    level3_00 += levels['3'][0]['initialGrade']
    level3_01 += levels['3'][0]['finalGrade']

    level3_10 += levels['3'][1]['initialGrade']
    level3_11 += levels['3'][1]['finalGrade']
    
    
    
```

```python
level1_00 = level1_00 / 20
level1_01 = level1_01 / 20

level1_10 = level1_10 / 20
level1_11 = level1_11 / 20

level1_20 = level1_20 / 20
level1_21 = level1_21 / 20

level1_30 = level1_30 / 20
level1_31 = level1_31 / 20

level1_40 = level1_40 / 20
level1_41 = level1_41 / 20

level2_00 = level2_00 / 20
level2_01 = level2_01 / 20

level2_10 = level2_10 / 20
level2_11 = level2_11 / 20


level3_00 = level3_00 / 20
level3_01 = level3_01 / 20

level3_10 = level3_10 / 20
level3_11 = level3_11 / 20


level1_0T = level1_0T / 20
level1_1T = level1_1T / 20
level1_2T = level1_2T / 20
level1_3T = level1_3T / 20
level1_4T = level1_4T / 20
```

```python
import matplotlib.pyplot as plt

plt.plot(["Module 1", "Module 2", "Module 3", "Module 4", "Module 5"], [level1_00,level1_10,level1_20,level1_30,level1_40],color='red', marker='o', label ='Initial Grade')
plt.plot(["Module 1", "Module 2", "Module 3", "Module 4", "Module 5"], [level1_01,level1_11,level1_21,level1_31,level1_41],color='blue', marker='o', label ='Final Grade')

plt.title('Level 1: Initial Grade Vs Final Grade')
plt.xlabel('Modules', fontsize=16)
plt.ylabel('Grades in %',fontsize=16)
plt.legend()
plt.show()
```

```python

plt.plot(["Module 1", "Module 2"], [level2_00,level2_10],color='red', marker='o', label ='Initial Grade')
plt.plot(["Module 1", "Module 2"], [level2_01,level2_11],color='blue', marker='o', label ='Final Grade')

plt.title('Level 2: Initial Grade Vs Final Grade')
plt.xlabel('Modules', fontsize=16)
plt.ylabel('Grades in %',fontsize=16)
plt.legend()
plt.show()
```

```python

plt.plot(["Module 1", "Module 2"], [level3_00,level3_10],color='red', marker='o', label ='Initial Grade')
plt.plot(["Module 1", "Module 2"], [level3_01,level3_11],color='blue', marker='o', label ='Final Grade')

plt.title('Level 3: Initial Grade Vs Final Grade')
plt.xlabel('Modules', fontsize=16)
plt.ylabel('Grades in %',fontsize=16)
plt.legend()
plt.show()
```

```python

plt.plot(["Module 1", "Module 2", "Module 3", "Module 4", "Module 5"], [level1_0T,level1_1T,level1_2T,level1_3T,level1_4T],color='green', marker='o', label =' Avg Time Spent On Flash Card')

plt.title('Average Amount Of Time Spent On Each Flash Card')
plt.xlabel('Modules', fontsize=16)
plt.ylabel('Time in seconds',fontsize=16)
plt.legend()
plt.show()
```
